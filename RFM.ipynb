```python
cust_orders['Order_date']=pd.to_datetime(cust_orders['Order_date'])
```

```python
from datetime import datetime
rfm_cust = cust_orders.groupby('Customer').agg({'Order_date': lambda x: (datetime.today() - x.max()).days, #Recency
                                                'Order_id': lambda x: len(x), #Frequency
                                                'Amount': lambda x: x.sum()}) #Monetary Value
```

```python
rfm_cust.head(5)
```

```python
cust_orders[cust_orders['Customer']=='Aakanksha']
```

```python
rfm_cust['Order_date'] = rfm_cust['Order_date'].astype(int)
rfm_cust.columns=['recency','frequency','monetary_value']
rfm_cust.head()
```

```python
rfm_cust['R_quartile']=pd.qcut(rfm_cust['recency'],q=4,labels=['1','2','3','4'])
```

```python
rfm_cust.head()
```

```python
rfm_cust['M_quartile']=pd.qcut(rfm_cust['monetary_value'],q=4,labels=['4','3','2','1'])
```

```python
rfm_cust.head()
```

```python
rfm_cust['F_quartile']=pd.qcut(rfm_cust['frequency'].rank(method='first'),q=4,labels=['4','3','2','1'])
```

```python
rfm_cust.head()
```

```python
rfmlist=[]
for i in range(0,len(rfm_cust)):
    rfmlist.append(str(rfm_cust.iloc[i]['R_quartile'])+str(rfm_cust.iloc[i]['F_quartile'])+str(rfm_cust.iloc[i]['M_quartile']))
rfm_cust['RFM']=rfmlist    
```
